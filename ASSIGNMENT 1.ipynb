# Write a python program to display all the header tags from wikipedia.org

- HEADER TAGS

```python
url = 'https://en.wikipedia.org/wiki/Main_Page'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
for header in header_tags:
    print(header.text)
```

# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url = 'https://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
name_and_year_of_release=[]
for i in soup.find_all('td',class_="titleColumn"):
    name_and_year_of_release.append(i.text)
name_and_year_of_release
for item in name_and_year_of_release:
    print(item.strip())
```

```python
rating=[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text)
rating
```

```python
for item in rating:
    print(item.strip())
```

```python
import pandas as pd
movies = []
table = soup.find('tbody', {'class': 'lister-list'})
for row in table.find_all('tr'):
    title = row.find('td', {'class': 'titleColumn'}).find('a').text
    rating = row.find('td', {'class': 'ratingColumn'}).find('strong').text
    year = row.find('td', {'class': 'titleColumn'}).find('span', {'class': 'secondaryInfo'}).text.strip('()')
    movies.append((title, rating, year))

df = pd.DataFrame(movies, columns=['Name', 'Rating', 'Year of Release'])
print(df.head(100))
```

# Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame

```python
url = "https://www.imdb.com/india/top-rated-indian-movies/"
response = requests.get(url)
html_content = response.content
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# find the list of movies
movies = soup.find_all("td", class_="titleColumn")
```

```python
movie_names = []
for movie in movies:
    # extract the movie name
    name = movie.find("a").get_text()
    movie_names.append(name)
```

```python
movie_names
```

```python
rating=[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text)
rating
```

```python
for item in rating:
    print(item.strip())
```

```python
year=[]
for i in soup.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
year
```

```python
df=pd.DataFrame({'indian movies':movie_names,'Rating':rating,'Year':year})
df
```

# Write a python program to scrape mentioned details from dineout.co.in :i) Restaurant nameii) Cuisineiii) Locationiv) Ratingsv) Image URL 

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
titles
```

```python
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
cuisine
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
ratings
```

```python
img=[]
for i in soup.find_all('img',class_="no-img"):
    img.append(i.get('data-src'))# instead of text we will use i.get method for image
img
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':titles,'Location':location,'Cuisine':cuisine,'IMAGE':img,'Ratings':ratings})
df
```
